```python
import cv2
import numpy as np
```

```python
#simple image with 3 black squares
image = cv2.imread('/content/edgedetec.jpg')
```

```python
#Grayscaling
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
```

```python
# Finding Canny edges
edged = cv2.Canny(gray, 30, 200)
```

```python
# Finding Contours uisng a copy of the image e.g. edged.copy()
# since findContours alters the image
contours, hierarchy = cv2.findContours(edged,
    cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_NONE)
```

```python
from google.colab.patches import cv2_imshow

# Assuming 'edged' is your image array
cv2_imshow(edged)

#cv2.imshow('Canny Edges After Contouring', edged) (disable on Colab)

print("\nNumber of Contours found = " + str(len(contours)))
```

```python
# Draw all contours
# -1 signifies drawing all contours
cv2.drawContours(image, contours, -1, (0, 255, 0), 3)
```
